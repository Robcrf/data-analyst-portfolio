3.3 Busca las plataformas que solían ser populares pero que ahora no tienen ventas. ¿Cuánto tardan generalmente las nuevas plataformas en aparecer y las antiguas en desaparecer? 

3.4 Determina para qué período debes tomar datos. Para hacerlo mira tus respuestas a las preguntas anteriores. Los datos deberían permitirte construir un modelo para 2017.  

3.5 Trabaja solo con los datos que consideras relevantes. Ignora los datos de años anteriores.  


```python

duracion_plataforma = df[df['year_of_release'] != 0].groupby('platform').agg(
    primer_ano = ('year_of_release','min'),
    ultimo_ano = ('year_of_release','max'),
    total_venta = ('total_sales', 'sum')
).reset_index()

duracion_plataforma['duracion'] = duracion_plataforma['ultimo_ano'] - duracion_plataforma['primer_ano']

plataformas_declive = duracion_plataforma[duracion_plataforma['ultimo_ano'] < 2016].sort_values(by='total_venta', ascending=False)

plataformas_declive
```

```python
duracion_promedio = plataformas_declive['duracion'].mean()

print(f"El promedio de años en que tardan en desaparecer estas plataformas en declive es de: {duracion_promedio} años")
```

```python
plataformas_nuevas = duracion_plataforma[duracion_plataforma['primer_ano'] >= 2012]

plataformas_nuevas
```

```python
generaciones_plataformas = df[df['platform'].str.contains('PS|^X|Wi')].groupby('platform').size()

lista_plataformas = generaciones_plataformas.index.to_list()

plataformas_generacionales = duracion_plataforma[duracion_plataforma['platform'].isin(lista_plataformas)]

plataformas_playstation = plataformas_generacionales[plataformas_generacionales['platform'].str.contains('PS')]['primer_ano'].sort_values(ascending=True).diff().mean()

plataformas_wii = plataformas_generacionales[plataformas_generacionales['platform'].str.contains('Wi')]['primer_ano'].sort_values(ascending=True).diff().mean()

plataformas_xbox = plataformas_generacionales[plataformas_generacionales['platform'].str.contains('^X')]['primer_ano'].sort_values(ascending=True).diff().mean()

promedio = (plataformas_playstation + plataformas_wii + plataformas_xbox) / len([plataformas_playstation,plataformas_wii, plataformas_xbox])

print(f"El promedio de años que tarda en salir una platafroma nueva generacional es de: {np.ceil(promedio)} años")
```

3.6 ¿Qué plataformas son líderes en ventas? ¿Cuáles crecen y cuáles se reducen? Elige varias plataformas potencialmente rentables.  

```python
plataformas_lideres =  duracion_plataforma[duracion_plataforma['primer_ano'] >= 2012].sort_values(by='total_venta',ascending=False)

lista = plataformas_lideres['platform'].to_list()

ventas_anuales_lideres = df[df['platform'].isin(lista)]

plt.figure(figsize=(12,6))
sns.lineplot(data=ventas_anuales_lideres, x='year_of_release', y='total_sales', hue='platform', style='platform',markers=True, dashes=False, estimator=sum)
plt.title('Tendencia de Ventas Anuales de Plataformas Líderes (2012 en adelante)', fontsize=16)
plt.xlabel('Año de Lanzamiento', fontsize=12)
plt.ylabel('Ventas Totales (millones)', fontsize=12)
plt.legend(title='Plataforma')
plt.xlim(2012, 2016) # Enfocar en el período relevante
plt.show()
```

```python
#Las plataformas de nueva generación son las que estan teniendo ventas fuertes PS4,XONE,WIIU, sin embargo las platafromas como PS2, DS,X360,PS3,WII 
# son plataformas que estan teniendo reducciones en ventas significativas en la actualidad llegando a ser 0.
```

## 5. Prueba las siguientes hipótesis

— Las calificaciones promedio de los usuarios para las plataformas Xbox One y PC son las mismas.  
— Las calificaciones promedio de los usuarios para los géneros de Acción y Deportes son diferentes.  

5.1 Establece tú mismo el valor de umbral alfa.  
5.2 Explica:  
&nbsp;&nbsp;&nbsp;&nbsp;• Cómo formulaste las hipótesis nula y alternativa.  
&nbsp;&nbsp;&nbsp;&nbsp;• Qué criterio utilizaste para probar las hipótesis y por qué.

```python
#Probar la hipótesis:
# H0: Las calificaciones promedio de los usuarios para las plataformas Xbox One y PC son las mismas.
# H1: Las calificaciones promedio de los usuarios para las plataformas Xbox One y PC son diferentes.
```

```python
#El nivel de significacia será de 0.05.
alpha = 0.05
```

```python
# Se calculan las calificaciones promedio de los usuarios para ambas plataformas.
promedio_xone = df_relevante[df_relevante['platform'] == 'XOne']['user_score'].mean()
pormedio_pc = df_relevante[df_relevante['platform'] == 'PC']['user_score'].mean()

print(f"Calificación promedio de usuarios en Xbox One: {promedio_xone:.2f}")
print(f"Calificación promedio de usuarios en PC: {pormedio_pc:.2f}")
```

```python
#Se calculan las varianzas de las calificaciones de usuarios para ambas plataformas.
varianza_xone = df_relevante[df_relevante['platform'] == 'XOne']['user_score'].var()
varianza_pc = df_relevante[df_relevante['platform'] == 'PC']['user_score'].var()

print(f"Varianza de calificaciones de usuarios en Xbox One: {varianza_xone:.2f}")
print(f"Varianza de calificaciones de usuarios en PC: {varianza_pc:.2f}")
```

```python
#Se calculan las varianzas usando el metodo de Levene para determinar si son iguales o no.
stat, p_value_levene = stats.levene(
    df_relevante[df_relevante['platform'] == 'XOne']['user_score'].dropna(),
    df_relevante[df_relevante['platform'] == 'PC']['user_score'].dropna()
)
print(f"Prueba de Levene - Estadístico: {stat:.2f}, p-valor: {p_value_levene:.4f}")

if p_value_levene < alpha:
    print("Las varianzas son significativamente diferentes.")
else:
    print("No hay evidencia suficiente para decir que las varianzas son diferentes.")
```

```python
#Se usan la prueba t de Student para muestras independientes para comparar las medias de las calificaciones de usuarios entre Xbox One y PC.
#En este caso, se usará la prueba con varianzas iguales o diferentes según el resultado de la prueba de Levene.
#Se usara el Twelchs t-test ya que las varianzas son diferentes, este es una variación de la prueba t que no asume que las 
#Varianzas son iguales entre los dos grupos,solo se asume que las varianzas son diferentes.
#Se usa el método dropna() para eliminar los valores nulos antes de realizar la prueba.
#Se usa el parámetro equal_var=False para indicar que las varianzas son diferentes.
#En este caso es una prueba de dos colas, ya que se quiere saber si las medias son diferentes, no se especifica una dirección.
#Se obtiene el estadístico t y el p-valor de la prueba.
#Se usa la prueba T de Welch para muestras inependientes por los siguientes motivos:
#1. Las varianzas de las calificaciones de usuarios entre Xbox One y PC son significativamente diferentes.
#2. La prueba T de Welch es más robusta cuando las varianzas son desiguales.
#3. La prueba T de Welch no asume varianzas iguales, lo que la hace más adecuada para este caso.
```

```python
#Se procede a usar la prueba T de Welch:
t_stat, p_value_test = stats.ttest_ind(
    df_relevante[df_relevante['platform'] == 'XOne']['user_score'].dropna(),
    df_relevante[df_relevante['platform'] == 'PC']['user_score'].dropna(),
    equal_var=False
)

p_value_test = p_value_test / 2  # Prueba de una cola

print(f"Prueba T de Welch - Estadístico t: {t_stat:.2f}, p-valor: {p_value_test:.4f}")

if p_value_test < alpha:
    print("Se rechaza la hipótesis nula: Las calificaciones promedio de los usuarios para Xbox One y PC son diferentes.")
else:
    print("No se rechaza la hipótesis nula: No hay evidencia suficiente para decir que las calificaciones promedio de los usuarios para Xbox One y PC son diferentes.")
```
